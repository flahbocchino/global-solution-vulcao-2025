```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Configura o gráfico para aparecer direto no notebook
%matplotlib inline
```

```python
# Simulando dados de temperatura e umidade
dados = {
    "tempo (s)": list(range(0, 200, 10)),  # tempo de 0 a 190, em passos de 10s
    "temperatura (°C)": [],
    "umidade (%)": []
}

for _ in dados["tempo (s)"]:
    # Valores dentro de faixas realistas (simulando o sensor DHT22)
    temp = random.uniform(20, 50)  # Temperatura entre 20°C e 50°C
    umid = random.uniform(10, 60)  # Umidade entre 10% e 60%
    dados["temperatura (°C)"].append(round(temp, 2))
    dados["umidade (%)"].append(round(umid, 2))

# Criar o DataFrame
df = pd.DataFrame(dados)
df.head()
```

```python
# Gráfico de temperatura e umidade
plt.figure(figsize=(12,5))

# Temperatura
plt.plot(df["tempo (s)"], df["temperatura (°C)"], label="Temperatura (°C)", color="red")

# Umidade
plt.plot(df["tempo (s)"], df["umidade (%)"], label="Umidade (%)", color="blue")

plt.axhline(45, color='red', linestyle='--', label='Limite de alerta (Temp > 45°C)')
plt.axhline(20, color='blue', linestyle='--', label='Limite de alerta (Umid < 20%)')

plt.title("Simulação de Condições Ambientais Próximas ao Vulcão")
plt.xlabel("Tempo (s)")
plt.ylabel("Medições")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Verificar condições críticas
print("⚠️ Verificando risco de erupção...\n")

for i in range(len(df)):
    temp = df.loc[i, "temperatura (°C)"]
    umid = df.loc[i, "umidade (%)"]
    tempo = df.loc[i, "tempo (s)"]

    if temp > 45 and umid < 20:
        print(f"⚠️ ALERTA no tempo {tempo}s - Temp: {temp}°C / Umidade: {umid}%")
```
